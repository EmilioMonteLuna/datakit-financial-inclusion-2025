---


- On July 28, 2022, staff from the National Institute of Public Health (INSP) began collecting field data for the 2022 Continuous National Health and Nutrition Survey (ENSANUT), which ran until December 10 of that year.
- ENSANUT Continua 2022 updates the overview of the frequency, distribution, and prevalence of major diseases and household health conditions, determines nutritional status and food consumption, monitors SARS-CoV-2 seroprevalence, updates COVID-19 vaccine coverage, identifies the presence of anemia and lead in the blood in the selected population, as well as other important indicators, and thereby accumulates information to be able to prepare state reports in 2024. 
- This is the third of five ongoing annual ENSANUTs; this time dedicated to continuing to research various aspects of COVID-19 on health and well-being, given the significant impact of the SARS-CoV-2 pandemic. 
- The main parameters of this survey are the presence of SARS-CoV-2 antibodies in blood serum, either through infection or vaccination; as well as vaccination coverage, prevalence of chronic diseases, and the general health and nutritional status of the population. It also provides an overview of lead exposure in children aged 1 to 4 years and pregnant women, as well as anemia in the selected population. A total of 14,240 homes were visited, obtaining complete information on approximately 10,450 households. These homes are distributed across eight regions of the country: North Pacific, Border, Central Pacific, North Central, Central, Mexico City-State of Mexico, South Pacific, and Peninsula. 
- To ensure the safety of both the surveyed population and the interviewers, procedures and a manual of protective measures were developed for field staff during training, at the start of the interview, and during the interview, as well as the actions to take if they experience COVID-19 symptoms during fieldwork. 
- ENSANUT Continua 2022 surveyors visited homes across the country, properly uniformed and identified, with protective equipment, and receiving the necessary training on hygiene and social distancing measures.
The National Health and Nutrition Surveys have a similar design, providing a comparable overview of the health of the Mexican population over time, with representation by urban and rural areas, by region, and by state. 
- Therefore, your participation is very important, as the information generated by this survey helps identify progress and challenges, as well as define health and nutrition priorities for the coming years. It also serves as a basis for assessing the medium- and long-term impact of the pandemic on population health and healthcare services.

```python

grouped = df.groupby(["desc_ent","h0101"]).agg(Count=("FOLIO_I", "count")).unstack()

label_map = {
    1: "Waste material 1",
    2: "Cardboard sheet 2",
    3: "Metal sheet",
    4: "Asbestos sheet",
    5: "Palm or straw",
    6: "Wood or shingles",
    7: "Roof with beams 7",
    8: "Tile 8",
    9: "Concrete slab or joists with vaults 9"
}

# Isolate the 'Count' level
grouped = grouped['Count']

# Rename columns
grouped.columns = grouped.columns.map(label_map)

# Convert to numeric and remove nulls
grouped = grouped.fillna(0)
#grouped = grouped.apply(pd.to_numeric, errors='coerce')

print(grouped)

# Plot
grouped.plot(kind='bar', stacked=True, figsize=(10,6))

plt.title("Survey Counts of Home Type (survey question H0101)")
plt.xlabel("state")
plt.ylabel("Count")
plt.legend(
    title="h0101",
    bbox_to_anchor=(1.05, 1),  # Position to the right
    loc='upper left',          # Anchor the upper left corner
    borderaxespad=0.
)
plt.tight_layout()
plt.show()
```
